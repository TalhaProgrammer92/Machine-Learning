# Scikit-Learn Usage

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

## Simple Linear Regression Model

```python

# ! Load data
data_frame = sb.load_dataset('tips')
data_frame.head()   # * Display the data
```

```python

# ! Regression line plot of the data frame
sb.lmplot(x='total_bill', y='tip', data=data_frame)
```

```python
import sklearn.linear_model as lm

# ! Feature column selection from the data frame
X = data_frame[['total_bill']]      # * Feature
y = data_frame[['tip']]             # * Label

# ! Load the model
model = lm.LinearRegression()

# ! Fit the model - Training
# model.fit(X.values.reshape(-1, 1), y)
model.fit(X, y)
```

```python

# ! Prediction of y (tip) based on X (total bill)
total_bill: int = 50
model.predict([[total_bill]])
```
